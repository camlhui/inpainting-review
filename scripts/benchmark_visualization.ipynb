```python
import json
import os


root_dir = '../'

with open(os.path.join(root_dir, 'inpainting-tasks.json')) as f:
    tasks = json.load(f)
```

```python
import textwrap
import matplotlib.pyplot as plt
%matplotlib inline


fig, axes = plt.subplots(nrows=len(tasks['tasks']), ncols=3, figsize=(15, 5 * len(tasks['tasks'])))

for i, task in enumerate(tasks['tasks']):
    # Display source image
    source_img = plt.imread(os.path.join(root_dir, task['source_image']))
    axes[i, 0].imshow(source_img)
    axes[i, 0].axis('off')
    axes[i, 0].set_title('Source Image')

    # Display mask image
    mask_img = plt.imread(os.path.join(root_dir, task['mask_image']))
    axes[i, 1].imshow(mask_img)
    axes[i, 1].axis('off')
    axes[i, 1].set_title('Mask Image')

    # Display prompt text
    wrapped_prompt = textwrap.fill(task['prompt'], width=50)
    axes[i, 2].text(0.5, 0.5, wrapped_prompt, ha='center', va='center', wrap=True)
    axes[i, 2].axis('off')
    axes[i, 2].set_title('Prompt')

plt.tight_layout()
plt.show()

fig.savefig(os.path.join(root_dir, 'data/outputs/benchmark.png'))
```
